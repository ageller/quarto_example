# Load in the data

Use the standard "iris" flower dataset from the [`seaborn`](https://seaborn.pydata.org/) package.

```python
#| label: load-data

import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

# Generate a corner plot using `seaborn` (@fig-iris)

```python
#| label: fig-iris
#| fig-cap: "Iris corner plot."

g = sns.PairGrid(iris, diag_sharey = False, hue = 'species', corner = True)
g.map_lower(sns.scatterplot)
g.map_diag(sns.kdeplot)
g.add_legend(bbox_to_anchor = (0.6, 0.6))
```
